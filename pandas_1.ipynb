

```
In Pandas, groupby() is a powerful method used to split a DataFrame into groups based on one or more criteria,
 apply a function to each group independently, and then combine the results into a new DataFrame or Series.
 This process is often referred to as the "split-apply-combine" strategy.
```



````python
xdf.groupby('category').sum()
````

````python
xdf.groupby('category').max()
````

````python
xdf.groupby('category').min()
````

````python
grouped=xdf.groupby('category').agg({'max','min','sum','mean'})
grouped
````

````python
grouped=xdf.groupby('category').agg({'values':['max','min','sum','mean']})
grouped
````

````python
# for each dno and its sub group gender find total salary
# sql :
# select dno, gender, sum(sal) from emp group by dno, gender
gtotal=df[['gender','dno','salary']]
gsala=gtotal.groupby(['dno','gender']).sum()
gsala
````

````python
# for each dno and its sub group gender find summary
# sql :
# select dno, gender, sum(sal),
# 	count(*) , avg(sal) , max(sal), min(sal)
#     from emp group by dno, gender
gtotall=df[['gender','dno','salary']]
gsummary=gtotall.groupby(['dno','gender']).agg(['count','max','min','sum','mean'])
gsummary
````
